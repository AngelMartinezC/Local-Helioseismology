```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
```

```python
def arc(frame=1,xrange=6,theta1=20,theta2=70,resolution=0.22,ax=None):

    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)
    A = []
    for i in np.arange(0.5,xrange,resolution):
        if i == np.arange(0.5,xrange,resolution)[frame]:
            B = patches.Arc((0,0), width=i, height=i, angle=0, theta1=theta1, 
                            theta2=theta2,linewidth=7, color='red')
        else:
            B = patches.Arc((0,0), width=i, height=i, angle=0, theta1=theta1, 
                            theta2=theta2,linewidth=1, color='black')
        A.append(B)
        
    for i in range(len(A)):
        ax.add_patch(A[i])
    
    return None
```

```python
def td(frame,xsunq,xrange,yrange=10,resolution=0.2, ax=None):
    
    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)
    F1 = np.arange(0.5,yrange,resolution)
    AA = []
    #for j in np.arange(0.5,F1[frame],resolution):
    j = F1[frame]
    for i in np.arange(0.5,xrange,resolution):
        arco = patches.Rectangle((i,j),width=resolution,height=resolution,
                                 linewidth=0.5,fill=False)
        X = np.arange(0.5,xrange,resolution)
        if i == X[xsunq]:
            arco = patches.Rectangle((i,j),width=resolution,linewidth=2,
                                     height=resolution,fill=True,color='red')
        AA.append(arco)    
    
    for i in range(len(AA)):
        ax.add_patch(AA[i])
    
    return None
```

```python
frame = 24

fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(121)
ax.set_xlim(-0.5,3)
ax.set_ylim(-0.5,3)
arc(frame,ax=ax)

xsunq = 4
xrange = 20
resolution = 0.2

ax = fig.add_subplot(122)
ax.set_xlim(0,7)
ax.set_ylim(0,7)

for i in range(3,26):
    AA = td(i,xsunq=int(np.exp(i/7.8)),xrange=xrange,ax=ax)
    if i == range(1,26)[-3]:
        AA = td(i+1,xsunq=20,xrange=xrange,ax=ax)
        AA = td(i,xsunq=18,xrange=xrange,ax=ax)
        AA = td(i+2,xsunq=22,xrange=xrange,ax=ax)
        AA = td(i+2,xsunq=23,xrange=xrange,ax=ax)
        AA = td(i+2,xsunq=24,xrange=xrange,ax=ax)
```
